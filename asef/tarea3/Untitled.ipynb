```python
# ============================================================
# Tome un notebook que ya tenia que hacia algo similar se lo pase al claude y le dije oye compa cambiame esto a esto otro, mas alla de eso no lo revise mucho
# ============================================================
import os
import time
import requests
import pandas as pd
from pathlib import Path
import pyarrow

TMDB_TOKEN = os.getenv("TMDB_TOKEN", "TOKEN_AQUI")

HEADERS = {
    "Authorization": f"Bearer {TMDB_TOKEN}",
    "accept": "application/json",
}

CACHE_DIR = Path("./animation_cache")
CACHE_DIR.mkdir(exist_ok=True)

SHOWS_PARQUET = CACHE_DIR / "tmdb_shows.parquet"
IMDB_IDS_PARQUET = CACHE_DIR / "tmdb_imdb_ids.parquet"
RATINGS_FILE = CACHE_DIR / "title.ratings.tsv.gz"

YEAR_START = 1960
YEAR_END = 2026
ANIMATION_GENRE_ID = 16
```

```python
# ============================================================
# Cell 2 — Pull animated shows from TMDb Discover (resumable + retries)
# ============================================================
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
CACHE_DIR.mkdir(parents=True, exist_ok=True)
SHOWS_PROGRESS_PARQUET = CACHE_DIR / "tmdb_shows_progress.parquet"
SHOWS_DONE_YEARS = CACHE_DIR / "tmdb_shows_done_years.txt"

def make_session():
    s = requests.Session()
    retry = Retry(
        total=8,
        backoff_factor=1.5,        # 1.5s, 3s, 6s, 12s, ...
        status_forcelist=[429, 500, 502, 503, 504],
        allowed_methods=["GET"],
        raise_on_status=False,
    )
    adapter = HTTPAdapter(max_retries=retry, pool_connections=10, pool_maxsize=10)
    s.mount("https://", adapter)
    s.mount("http://", adapter)
    s.headers.update(HEADERS)
    return s

def load_done_years():
    if not SHOWS_DONE_YEARS.exists():
        return set()
    try:
        return set(int(x) for x in SHOWS_DONE_YEARS.read_text().split() if x.strip())
    except FileNotFoundError:
        return set()

def mark_year_done(year):
    with open(SHOWS_DONE_YEARS, "a") as f:
        f.write(f"{year}\n")

def fetch_year(session, year, max_attempts=5):
    """Fetch all pages for a given year, with manual retry on connection errors."""
    rows = []
    page = 1
    while True:
        for attempt in range(max_attempts):
            try:
                r = session.get(
                    "https://api.themoviedb.org/3/discover/tv",
                    params={
                        "with_genres": ANIMATION_GENRE_ID,
                        "first_air_date_year": year,
                        "page": page,
                        "sort_by": "popularity.desc",
                    },
                    timeout=20,
                )
                if r.status_code == 429:
                    time.sleep(3)
                    continue
                r.raise_for_status()
                data = r.json()
                break
            except (requests.ConnectionError, requests.Timeout) as e:
                wait = 2 ** attempt
                print(f"    year {year} page {page} attempt {attempt+1} failed ({e.__class__.__name__}), sleeping {wait}s")
                time.sleep(wait)
        else:
            raise RuntimeError(f"Year {year} page {page} failed after {max_attempts} attempts")

        for show in data.get("results", []):
            rows.append({
                "tmdb_id": show["id"],
                "name": show.get("name"),
                "first_air_date": show.get("first_air_date"),
                "origin_country": ",".join(show.get("origin_country", []) or []),
                "original_language": show.get("original_language"),
                "vote_average_tmdb": show.get("vote_average"),
                "vote_count_tmdb": show.get("vote_count"),
                "popularity": show.get("popularity"),
            })

        total_pages = min(data.get("total_pages", 1), 500)
        if page >= total_pages:
            break
        page += 1
        time.sleep(0.05)
    return rows

def fetch_shows_resumable():
    # If final cache exists, just use it
    if SHOWS_PARQUET.exists():
        print(f"Loading cached shows from {SHOWS_PARQUET}")
        return pd.read_parquet(SHOWS_PARQUET)

    # Load any partial progress
    if SHOWS_PROGRESS_PARQUET.exists():
        progress = pd.read_parquet(SHOWS_PROGRESS_PARQUET)
        all_rows = progress.to_dict("records")
        print(f"Resuming with {len(all_rows)} rows already collected")
    else:
        all_rows = []

    done_years = load_done_years()
    session = make_session()

    for year in range(YEAR_START, YEAR_END + 1):
        if year in done_years:
            continue
        try:
            year_rows = fetch_year(session, year)
            all_rows.extend(year_rows)
            mark_year_done(year)

            # Checkpoint after every year
            pd.DataFrame(all_rows).drop_duplicates("tmdb_id").to_parquet(
                SHOWS_PROGRESS_PARQUET, index=False
            )
            print(f"  {year}: +{len(year_rows)} rows, cumulative {len(all_rows)}")
        except Exception as e:
            print(f"  {year}: FAILED ({e}) — rerun the cell to resume")
            raise

    df = pd.DataFrame(all_rows).drop_duplicates("tmdb_id").reset_index(drop=True)
    df.to_parquet(SHOWS_PARQUET, index=False)
    print(f"Saved {len(df)} shows to {SHOWS_PARQUET}")
    return df

shows = fetch_shows_resumable()
shows.head()
```

```python
# ============================================================
# Cell 3 — Fetch IMDb IDs (multithreaded, resumable)
# Esta la tenia hecha pero se demoraba mucho con estos datos asi que le dije al claudio que me lo cambiara a multithread
# ============================================================
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading

IMDB_IDS_PKL = CACHE_DIR / "tmdb_imdb_ids.pkl"  # adjust if you kept .parquet

MAX_WORKERS = 16              # TMDb tolerates ~50 req/sec; 16 threads is safe
CHECKPOINT_EVERY = 500

def fetch_one_imdb_id(session, tmdb_id, max_attempts=5):
    for attempt in range(max_attempts):
        try:
            r = session.get(
                f"https://api.themoviedb.org/3/tv/{tmdb_id}/external_ids",
                timeout=15,
            )
            if r.status_code == 429:
                time.sleep(2 + attempt)
                continue
            if r.status_code == 404:
                return tmdb_id, None
            r.raise_for_status()
            return tmdb_id, r.json().get("imdb_id")
        except (requests.ConnectionError, requests.Timeout):
            time.sleep(2 ** attempt)
    return tmdb_id, None

def fetch_imdb_ids_parallel(shows_df):
    if IMDB_IDS_PKL.exists():
        cached = pd.read_pickle(IMDB_IDS_PKL)
    else:
        cached = pd.DataFrame(columns=["tmdb_id", "imdb_id"])

    done_ids = set(cached["tmdb_id"].tolist())
    todo = shows_df[~shows_df["tmdb_id"].isin(done_ids)]["tmdb_id"].tolist()
    print(f"{len(done_ids)} cached, {len(todo)} to fetch with {MAX_WORKERS} workers")

    if not todo:
        return cached

    session = make_session()  # from Cell 2; reuses connection pool, retries
    results = []
    lock = threading.Lock()
    completed = 0

    def checkpoint():
        combined = pd.concat(
            [cached, pd.DataFrame(results, columns=["tmdb_id", "imdb_id"])],
            ignore_index=True,
        ).drop_duplicates("tmdb_id")
        combined.to_pickle(IMDB_IDS_PKL)

    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as ex:
        futures = {ex.submit(fetch_one_imdb_id, session, tid): tid for tid in todo}
        for fut in as_completed(futures):
            tmdb_id, imdb_id = fut.result()
            with lock:
                results.append((tmdb_id, imdb_id))
                completed += 1
                print(f"Completed {fut}")
                if completed % CHECKPOINT_EVERY == 0:
                    checkpoint()
                    print(f"  {completed}/{len(todo)} done, checkpointed")

    checkpoint()
    combined = pd.read_pickle(IMDB_IDS_PKL)
    print(f"Total IMDb IDs cached: {len(combined)}")
    return combined

imdb_ids = fetch_imdb_ids_parallel(shows)
imdb_ids.head()
```

```python
# ============================================================
# Cell 4 — Load IMDb ratings
# Se explica solo
# ============================================================
def load_imdb_ratings():
    if not RATINGS_FILE.exists():
        print("Downloading IMDb ratings...")
        url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
        r = requests.get(url, stream=True, timeout=60)
        r.raise_for_status()
        with open(RATINGS_FILE, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Saved to {RATINGS_FILE}")

    ratings = pd.read_csv(
        RATINGS_FILE,
        sep="\t",
        usecols=["tconst", "averageRating", "numVotes"],
    ).rename(columns={"tconst": "imdb_id"})
    print(f"Loaded {len(ratings):,} IMDb ratings")
    return ratings

ratings = load_imdb_ratings()
ratings.head()
```

```python
# ============================================================
# Cell 5 — Join & clean
#Adaptacion bien literal de lo que ya tenia
# ============================================================
MIN_VOTES = 1000

df = (
    shows
    .merge(imdb_ids, on="tmdb_id", how="left")
    .merge(ratings, on="imdb_id", how="inner")
)

# Take the first listed country (simplification; revisit for co-prods later)
df["country"] = df["origin_country"].str.split(",").str[0]
df = df[df["country"].notna() & (df["country"] != "")]

df["year"] = pd.to_datetime(df["first_air_date"], errors="coerce").dt.year

df_filtered = df[df["numVotes"] >= MIN_VOTES].copy()
print(f"{len(df)} merged rows, {len(df_filtered)} after vote threshold")
df_filtered.head()

```

```python
# ============================================================
# Cell 6 — Country ranking (Bayesian-weighted)
# ============================================================
MIN_SHOWS_PER_COUNTRY = 20
PRIOR_WEIGHT = 500  # m: tune higher to pull more toward global mean

C = df_filtered["averageRating"].mean()  # global mean rating

df_filtered["weighted_rating"] = (
    df_filtered["numVotes"] * df_filtered["averageRating"] + PRIOR_WEIGHT * C
) / (df_filtered["numVotes"] + PRIOR_WEIGHT)

ranking = (
    df_filtered.groupby("country")
    .agg(
        n_shows=("imdb_id", "count"),
        avg_rating=("averageRating", "mean"),
        weighted_rating=("weighted_rating", "mean"),
        total_votes=("numVotes", "sum"),
    )
    .query(f"n_shows >= {MIN_SHOWS_PER_COUNTRY}")
    .sort_values("weighted_rating", ascending=False)
    .round(3)
)

print(f"Global mean rating: {C:.3f}")
ranking
```

```python
# ============================================================
# Cell 6b
# Le cambie las cantidades de votos y cosas minimas para que aparezcan mas en el mapa
# ============================================================
MIN_VOTES_MAP = 100         # 1000
MIN_SHOWS_MAP = 3           #  20
PRIOR_WEIGHT_MAP = 200      #  500 

df_map = df[df["numVotes"] >= MIN_VOTES_MAP].copy()
C_map = df_map["averageRating"].mean()

df_map["weighted_rating"] = (
    df_map["numVotes"] * df_map["averageRating"] + PRIOR_WEIGHT_MAP * C_map
) / (df_map["numVotes"] + PRIOR_WEIGHT_MAP)

ranking_map = (
    df_map.groupby("country")
    .agg(
        n_shows=("imdb_id", "count"),
        avg_rating=("averageRating", "mean"),
        weighted_rating=("weighted_rating", "mean"),
        total_votes=("numVotes", "sum"),
    )
    .query(f"n_shows >= {MIN_SHOWS_MAP}")
    .sort_values("weighted_rating", ascending=False)
    .round(3)
)


print(f"Loose ranking:  {len(ranking_map)} countries")
ranking_map.head(20)
```

```python
# ============================================================
# Cell 7 — Export results to readable files
# Claudio Original, es solo para guardar los archivos para subir al git, quedan en csv, o si quieren tambien en excel pero no me di la lata
# ============================================================
OUTPUT_DIR = Path("./animation_output")
OUTPUT_DIR.mkdir(exist_ok=True)

# 1. Country ranking — the main result
ranking_out = ranking.reset_index()
ranking_out.to_csv(OUTPUT_DIR / "country_ranking.csv", index=False)

# 2. Full joined dataset — every show with country, year, ratings
shows_out = df_filtered[[
    "tmdb_id", "imdb_id", "name", "country", "origin_country",
    "year", "first_air_date", "original_language",
    "averageRating", "numVotes", "weighted_rating",
    "vote_average_tmdb", "vote_count_tmdb", "popularity",
]].sort_values(["country", "weighted_rating"], ascending=[True, False])
shows_out.to_csv(OUTPUT_DIR / "shows_full.csv", index=False)

# 4. Excel workbook with everything in one file (optional, nice for sharing)
try:
    with pd.ExcelWriter(OUTPUT_DIR / "animation_analysis.xlsx", engine="openpyxl") as xl:
        ranking_out.to_excel(xl, sheet_name="Country Ranking", index=False)
        shows_out.to_excel(xl, sheet_name="All Shows", index=False)
        ranking.sort_values("n_shows", ascending=False).reset_index().to_excel(
            xl, sheet_name="By Volume", index=False
        )
    print("Excel workbook written.")
except ImportError:
    print("openpyxl not installed — skipping .xlsx export. Run: pip install openpyxl")

print(f"\nFiles written to {OUTPUT_DIR.resolve()}:")
for f in sorted(OUTPUT_DIR.iterdir()):
    print(f"  {f.name}  ({f.stat().st_size:,} bytes)")
```

```python
# ============================================================
# Cell B — Co-production network (static)
# ============================================================
# pip install networkx matplotlib pycountry

import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter
import pycountry

MIN_JOINT_SHOWS = 5

# ---- 1. Build country pairs from co-productions ----
pair_counts = Counter()
country_show_counts = Counter()

for countries_str in df["origin_country"]:
    countries = [c for c in countries_str.split(",") if c]
    for c in countries:
        country_show_counts[c] += 1
    # All unordered pairs of co-producers on this show
    for a, b in combinations(sorted(set(countries)), 2):
        pair_counts[(a, b)] += 1

print(f"Total country pairs: {len(pair_counts)}")
print(f"Pairs with >= {MIN_JOINT_SHOWS} joint shows: "
      f"{sum(1 for v in pair_counts.values() if v >= MIN_JOINT_SHOWS)}")

# ---- 2. Build the graph ----
G = nx.Graph()
for (a, b), w in pair_counts.items():
    if w >= MIN_JOINT_SHOWS:
        G.add_edge(a, b, weight=w)

# Drop isolated nodes (countries with no qualifying co-prods)
G.remove_nodes_from(list(nx.isolates(G)))
print(f"Graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

# ---- 3. Node attributes for sizing/labeling ----
def country_name(code):
    try:
        return pycountry.countries.get(alpha_2=code).name
    except (AttributeError, KeyError):
        return code

node_sizes = [country_show_counts[n] for n in G.nodes()]
# Scale to reasonable visual range (sqrt to compress extremes)
max_shows = max(node_sizes)
node_sizes_scaled = [300 + 3000 * (s / max_shows) ** 0.5 for s in node_sizes]

edge_weights = [G[u][v]["weight"] for u, v in G.edges()]
max_edge = max(edge_weights)
edge_widths = [0.5 + 6 * (w / max_edge) ** 0.7 for w in edge_weights]

labels = {n: country_name(n) for n in G.nodes()}

# ---- 4. Layout: spring layout, weighted so heavy edges pull nodes closer ----
pos = nx.spring_layout(G, weight="weight", k=1.2, iterations=200, seed=42)

# ---- 5. Draw ----
fig, ax = plt.subplots(figsize=(16, 12))

nx.draw_networkx_edges(
    G, pos,
    width=edge_widths,
    edge_color=edge_weights,
    edge_cmap=plt.cm.plasma,
    alpha=0.6,
    ax=ax,
)
nx.draw_networkx_nodes(
    G, pos,
    node_size=node_sizes_scaled,
    node_color="#4A90E2",
    edgecolors="white",
    linewidths=1.5,
    alpha=0.9,
    ax=ax,
)
nx.draw_networkx_labels(
    G, pos,
    labels=labels,
    font_size=9,
    font_weight="bold",
    ax=ax,
)

# Colorbar for edge weight
sm = plt.cm.ScalarMappable(
    cmap=plt.cm.plasma,
    norm=plt.Normalize(vmin=min(edge_weights), vmax=max(edge_weights)),
)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.5, label="Joint productions")

ax.set_title(
    f"Animated Show Co-Production Network\n"
    f"(node size = total shows, edge width/color = joint productions, min {MIN_JOINT_SHOWS})",
    fontsize=14,
)
ax.axis("off")
plt.tight_layout()

OUTPUT_DIR = Path("./animation_output")
OUTPUT_DIR.mkdir(exist_ok=True)
plt.savefig(OUTPUT_DIR / "coproduction_network.png", dpi=200, bbox_inches="tight")
plt.show()

# ---- 6. Print the top 15 pairs as a sanity check ----
top_pairs = sorted(pair_counts.items(), key=lambda x: -x[1])[:15]
print("\nTop 15 co-production pairs:")
for (a, b), n in top_pairs:
    print(f"  {country_name(a):20s} <-> {country_name(b):20s} {n:4d}")
```

```python
# ============================================================
# Cell B — Co-production network 
# ============================================================
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter
import pycountry

MIN_JOINT_SHOWS = 5
MIN_NODE_DEGREE = 2   # drop countries with fewer than N connections

# ---- 1. Build pairs ----
pair_counts = Counter()
country_show_counts = Counter()

for countries_str in df["origin_country"]:
    countries = [c for c in countries_str.split(",") if c]
    for c in countries:
        country_show_counts[c] += 1
    for a, b in combinations(sorted(set(countries)), 2):
        pair_counts[(a, b)] += 1

# ---- 2. Build graph + prune ----
G = nx.Graph()
for (a, b), w in pair_counts.items():
    if w >= MIN_JOINT_SHOWS:
        G.add_edge(a, b, weight=w)

# Drop weakly connected nodes
weak_nodes = [n for n, d in G.degree() if d < MIN_NODE_DEGREE]
G.remove_nodes_from(weak_nodes)
G.remove_nodes_from(list(nx.isolates(G)))
print(f"Graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

# ---- 3. Node + edge styling ----
def country_name(code):
    overrides = {"KR": "South Korea"}
    if code in overrides:
        return overrides[code]
    try:
        return pycountry.countries.get(alpha_2=code).name
    except (AttributeError, KeyError):
        return code
# Sort nodes by total show count so the circular layout is ordered
nodes_sorted = sorted(G.nodes(), key=lambda n: -country_show_counts[n])
max_shows = max(country_show_counts[n] for n in G.nodes())
node_sizes = [400 + 2500 * (country_show_counts[n] / max_shows) ** 0.6 for n in nodes_sorted]

edge_weights = [G[u][v]["weight"] for u, v in G.edges()]
max_edge = max(edge_weights)
# Edge width directly encodes weight — clear and self-explanatory
edge_widths = [0.8 + 7 * (w / max_edge) for w in edge_weights]
# Edge opacity also scales with weight so weak ties recede visually
edge_alphas = [0.25 + 0.55 * (w / max_edge) for w in edge_weights]

# ---- 4. Circular layout, nodes ordered by volume ----
pos = nx.circular_layout(G.subgraph(nodes_sorted))
# circular_layout uses the order of G.nodes(); force our sorted order:
import numpy as np
n = len(nodes_sorted)
pos = {
    node: np.array([np.cos(2 * np.pi * i / n), np.sin(2 * np.pi * i / n)])
    for i, node in enumerate(nodes_sorted)
}

# ---- 5. Draw ----
fig, ax = plt.subplots(figsize=(13, 13))
fig.patch.set_facecolor("white")

# Edges first (so nodes overlay them)
for (u, v), w, width, alpha in zip(G.edges(), edge_weights, edge_widths, edge_alphas):
    x = [pos[u][0], pos[v][0]]
    y = [pos[u][1], pos[v][1]]
    ax.plot(x, y, color="#2C3E50", linewidth=width, alpha=alpha, zorder=1,
            solid_capstyle="round")

# Nodes
nx.draw_networkx_nodes(
    G, pos, nodelist=nodes_sorted,
    node_size=node_sizes,
    node_color="#E74C3C",
    edgecolors="white",
    linewidths=2,
    ax=ax,
)

# Labels positioned outside the circle for readability
label_pos = {
    node: np.array([np.cos(2*np.pi*i/n)*1.18, np.sin(2*np.pi*i/n)*1.18])
    for i, node in enumerate(nodes_sorted)
}
for node, (x, y) in label_pos.items():
    angle = np.degrees(np.arctan2(y, x))
    ha = "left" if x > 0.05 else "right" if x < -0.05 else "center"
    ax.text(x, y, country_name(node), fontsize=11, fontweight="bold",
            ha=ha, va="center", color="#2C3E50")

ax.set_title(
    "Red de Co-Produccion de Shows Animados",
    fontsize=18, fontweight="bold", pad=20, color="#2C3E50",
)
ax.text(0.5, -0.02,
        f"Tamaño del Nodo = Total de Shows · Grosor Conexion = Co-Producciones · Min {MIN_JOINT_SHOWS} compartidas",
        transform=ax.transAxes, ha="center", fontsize=10, color="#7F8C8D", style="italic")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.4, 1.4)
ax.set_aspect("equal")
ax.axis("off")
plt.tight_layout()

plt.savefig(OUTPUT_DIR / "coproduction_network.png", dpi=200, bbox_inches="tight",
            facecolor="white")
plt.show()
```

```python
# ============================================================
# Cell B — Co-production network
# ============================================================
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from itertools import combinations
from collections import Counter
import pycountry
import matplotlib.font_manager as fm
fm.fontManager.addfont("./animation_output/fonts/DMSerifDisplay-Regular.ttf")
fm.fontManager.addfont("./animation_output/fonts/DMSerifDisplay-Italic.ttf")
fm.fontManager.addfont("./animation_output/fonts/RedditSans-VariableFont_wght.ttf")
fm.fontManager.addfont("./animation_output/fonts/RedditSans-Italic-VariableFont_wght.ttf")
# ============================================================
# VISUAL LANGUAGE
# ============================================================
PALETTE = {
    "bg":        "#FAF7F2",   # fondo blanco crema
    "primary":   "#FF006E",   # nodos
    "secondary": "#3A0CA3",   # edges
    "accent":    "#E9C46A",   # highlitghts
    "muted":     "#5C5C5C",   # subtitulos
}

TYPOGRAPHY = {
    "title_font":    "Reddit Sans",     
    "body_font":     "Reddit Sans",
    "title_size":    28,
    "subtitle_size": 16,
    "label_size":    16,
}

LINES = {
    "thin":  1,
    "med":   3,
    "thick": 6,
}

NODE_STROKE_WIDTH = 2.5   # white outline around each node (sticker effect)
LABEL_RADIUS      = 1.18  # how far outside the circle labels sit

# ============================================================
# DATA PARAMETERS
# ============================================================
MIN_JOINT_SHOWS  = 5
MIN_NODE_DEGREE  = 2

# ---- 1. Build pairs ----
pair_counts = Counter()
country_show_counts = Counter()

for countries_str in df["origin_country"]:
    countries = [c for c in countries_str.split(",") if c]
    for c in countries:
        country_show_counts[c] += 1
    for a, b in combinations(sorted(set(countries)), 2):
        pair_counts[(a, b)] += 1

# ---- 2. Build & prune graph ----
G = nx.Graph()
for (a, b), w in pair_counts.items():
    if w >= MIN_JOINT_SHOWS:
        G.add_edge(a, b, weight=w)
G.remove_nodes_from([n for n, d in G.degree() if d < MIN_NODE_DEGREE])
G.remove_nodes_from(list(nx.isolates(G)))
print(f"Graph: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

# ---- 3. Country name helper (with overrides) ----
def country_name(code):
    overrides = {
        "KR": "South Korea", "KP": "North Korea", "RU": "Russia",
        "TW": "Taiwan", "VN": "Vietnam", "IR": "Iran", "CZ": "Czechia",
    }
    if code in overrides:
        return overrides[code]
    try:
        return pycountry.countries.get(alpha_2=code).name
    except (AttributeError, KeyError):
        return code

# ---- 4. Sort nodes by volume; compute sizes & edge styles ----
nodes_sorted = sorted(G.nodes(), key=lambda n: -country_show_counts[n])
max_shows = max(country_show_counts[n] for n in G.nodes())
node_sizes = [500 + 2800 * (country_show_counts[n] / max_shows) ** 0.6 for n in nodes_sorted]

edge_weights = [G[u][v]["weight"] for u, v in G.edges()]
max_edge = max(edge_weights)
edge_widths = [LINES["thin"] + (LINES["thick"] - LINES["thin"]) * (w / max_edge) for w in edge_weights]
edge_alphas = [0.20 + 0.60 * (w / max_edge) for w in edge_weights]

# ---- 5. Circular layout, ordered by volume ----
n = len(nodes_sorted)
pos = {
    node: np.array([np.cos(2 * np.pi * i / n), np.sin(2 * np.pi * i / n)])
    for i, node in enumerate(nodes_sorted)
}

# ---- 6. Draw ----
fig, ax = plt.subplots(figsize=(13, 13))
fig.patch.set_facecolor(PALETTE["bg"])
ax.set_facecolor(PALETTE["bg"])
# ---- Paper texture background ----
np.random.seed(42)  # so the texture is stable across runs

# Edges (under nodes)
for (u, v), width, alpha in zip(G.edges(), edge_widths, edge_alphas):
    x = [pos[u][0], pos[v][0]]
    y = [pos[u][1], pos[v][1]]
    ax.plot(x, y,
            color=PALETTE["secondary"],
            linewidth=width,
            alpha=alpha,
            zorder=1,
            solid_capstyle="round")

# Nodes
nx.draw_networkx_nodes(
    G, pos, nodelist=nodes_sorted,
    node_size=node_sizes,
    node_color=PALETTE["primary"],
    edgecolors=PALETTE["bg"],
    linewidths=NODE_STROKE_WIDTH,
    ax=ax,
)

# Labels outside the circle
for i, node in enumerate(nodes_sorted):
    angle = 2 * np.pi * i / n
    x, y = np.cos(angle) * LABEL_RADIUS, np.sin(angle) * LABEL_RADIUS
    ha = "left" if x > 0.05 else "right" if x < -0.05 else "center"
    ax.text(x, y, country_name(node),
            fontsize=TYPOGRAPHY["label_size"],
            fontfamily=TYPOGRAPHY["body_font"],
            fontweight="bold",
            ha=ha, va="center",
            color=PALETTE["secondary"])

# Title + subtitle
ax.text(0, 1.42, "Red de Co-Produccion de Shows Animados",
        fontsize=TYPOGRAPHY["title_size"],
        fontfamily=TYPOGRAPHY["title_font"],
        fontweight="bold",
        ha="center", color=PALETTE["secondary"])

ax.text(0, 1.34,
        f"Tamaño del Nodo = Total de Shows · Grosor Conexion = Co-Producciones · Min {MIN_JOINT_SHOWS} compartidas",
        fontsize=TYPOGRAPHY["subtitle_size"],
        fontfamily=TYPOGRAPHY["body_font"],
        style="italic",
        ha="center", color=PALETTE["muted"])

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.55)
ax.set_aspect("equal")
ax.axis("off")
plt.tight_layout()

plt.savefig(OUTPUT_DIR / "coproduction_network_reddit_gruesa_grisaseo.png",
            dpi=200, bbox_inches="tight",
            facecolor=PALETTE["bg"])
plt.show()
```
